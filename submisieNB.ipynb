```python
import os
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
import pandas as pd
data_path = ''
train_data_df = pd.read_csv(os.path.join(data_path, 'train_data.csv'))
train_data_df
```

```python
## Codificarea etichetelor din string in int
```

```python
#Exercitiu: codificam etichetele / labels in valori cu numere intregi dela 0 la N
# codificam etichetele / labels in valori cu numere intregi dela 0 la N
# ar putea fi exercitiu la clasa ca sa se obisnuiasca cu dictionare
etichete_unice = train_data_df['label'].unique()
print(etichete_unice)
label2id = {}
id2label = {}
for idx, eticheta in enumerate(etichete_unice):
    label2id[eticheta] = idx
    id2label[idx] = eticheta

print(label2id)
print(id2label)
```

```python
y_labels = []
for eticheta in train_data_df['label']:
    y_labels.append(label2id[eticheta])
y_labels = np.array(y_labels)

print(y_labels[:100])
```

```python
## Pre-procesarea datelor
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

exemple_italian = train_data_df[train_data_df['language'] == 'italiano']
print(exemple_italian)
```

```python
def proceseaza(text):
    text = re.sub("[-–.,;:!\’?\"\'\/()_*=`]", "", text)
    text = text.lower()
    text = word_tokenize(text)
    return text
```

```python
exemplu = "– Señor Pres’idente, doy las gracias al Comisario, no solo por venir hoy aquí a escuchar el debate, sino también por la ayuda que creo que prestó a los diputados en Ginebra la semana pasada. No pude estar allí, pero ellos expresaron su agradecimiento por la forma en que los mantuvo informados."
exemplu = re.sub("[-–.,;:!’?\"\'\/()_*=`]", "", exemplu)
exemplu = word_tokenize(exemplu)

exemplu
```

```python
# Exercitiu: aplicati functia pe fiecare exemplu de antrenare
#data = []
data = train_data_df['text'].apply(lambda text: proceseaza(text))
print(data[33255])
print(type(data))
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data, y_train, y_test = train_test_split(data, y_labels, test_size=0.33, random_state=42)

train_data, valid_data, y_train, y_valid = train_test_split(train_data, y_train, test_size=0.33, random_state = 42)
```

```python
train_data2 = train_data.map(' '.join)
test_data2= test_data.map(' '.join)
valid_data2 = valid_data.map(' '.join)

train_data2
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features = 75000, tokenizer = lambda x: x, preprocessor = lambda x: x, max_df= 0.25)

vectorizer.fit(train_data)

print(vectorizer.vocabulary_)
```

```python
X_train = vectorizer.transform(train_data)
X_valid = vectorizer.transform(valid_data)
X_test = vectorizer.transform(test_data)
```

```python
from sklearn import preprocessing

X_train = preprocessing.normalize(X_train, axis = 1)
X_valid = preprocessing.normalize(X_valid, axis = 1)
X_test = preprocessing.normalize(X_test, axis = 1)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV

model = MultinomialNB(alpha=0.5, fit_prior=True)

param={'alpha': [0.00001, 0.0001, 0.001, 0.1, 1, 10, 100,1000]}
    
clf=GridSearchCV(model,param,cv=10,return_train_score=True)  

print('Fitting...')
clf.fit(X_train, y_train)

print('Predicting...')
y_pred = clf.predict(X_valid)
print((y_pred == y_valid).mean()*100)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_pred,y_valid))
```
